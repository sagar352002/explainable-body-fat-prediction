## Body Fat Dataset

## ğŸ“Œ What this dataset is about?

This dataset have measurements of **252 men**. The main goal is to understand how body fat can be predicted using simple body measurments like abdomen size, chest size, weight, height etc.

The dataset also include **body density**, which is measured using underwater weighing. This method is very accurate but too hard and costly to do for normal people. So these data is used to create easier prediction formulas for body fat.

---

## ğŸ§ª What variables are inside?

Here is what each column means in simple words:

- **Density** â€“ body density measured by underwater weighing. This is like the gold standard.
- **BodyFat** â€“ percent of body fat calculated using Siri equation (495/D â€“ 450).
- **Age** â€“ age of the person (years).
- **Weight** â€“ weight in lbs.
- **Height** â€“ height in inches.

### ğŸ“ Body circumference (all in cm)

These are the body part sizes measured using tape:

- **Neck**
- **Chest**
- **Abdomen2** â€“ around the belly at umbilicus level
- **Hip**
- **Thigh**
- **Knee**
- **Ankle**
- **Biceps**
- **Forearm**
- **Wrist**

These measurement standards are taken from Behnke and Wilmore book (1974). Abdomen2 is measured "laterally at the iliac crest and anteriorly at the belly button".

---

## ğŸ¤” Why this dataset was created?

Many fitness books and researchers try to estimate body fat without using expensive tools. Underwater weighing is accurate but not possible for common people.

So researchers used these 252 cases to build **prediction equations** for body fat using simple measurements. One well-known paper using this dataset is from Penrose, Nelson and Fisher (1985).

---

## ğŸ” How body fat is calculated (simple version)

Scientists assume the body have two parts:
- lean tissue  
- fat tissue  

Using density, they can estimate fat percentage with Siri's formula:

\[
\text{Body Fat %} = 495 / \text{Density} - 450
\]

This dataset gives both the density and body fat % computed from this formula.

In Multiple Linear Regression (MLR), the assumption regarding **multicollinearity** is basically:

---

### **ğŸ“Œ Assumption of Multicollinearity in MLR**

> **Independent variables should not be highly correlated with each other.**

This means every predictor should contribute **unique information** to the model, rather than repeating the same information as another variable.

---

### **ğŸ§  Why this assumption exists?**

If two features are highly correlated (e.g., Weight and Abdomen), the model **can't decide which variable actually affects the target**, because both move together.

This causes:

* Large swings in coefficients
* Wrong interpretation
* High standard errors
* Unstable model when new data arrives

---

### **ğŸ“Œ When Multicollinearity is a Problem?**

| VIF Value | Problem Level                                   |
| --------- | ----------------------------------------------- |
| 1â€“5       | OK (No issue)                                   |
| 5â€“10      | Moderate multicollinearity                      |
| >10       | Serious multicollinearity â†’ **Remove variable** |

---

### **ğŸ“Œ Symptoms of Multicollinearity**

* Model has high **RÂ²**, but coefficients are insignificant
* Coefficients change drastically when adding/removing features
* Variables highly correlated with each other (|corr| > 0.7)
* High **VIF values**

---

### **ğŸ“Œ Why it's a problem (Intuition)**

Example:

```
Weight â‰ˆ Chest â‰ˆ Abdomen â‰ˆ Hip
```

All represent body size â†’ giving same signal â†’ model gets confused which one actually predicts body fat.

It doesn't break prediction performance, but breaks interpretation.

---

### **ğŸ“Œ Quick Checklist (Notebook Format)**

```
MLR assumes:
âœ” Predictors not strongly correlated with each other
âœ” Each variable contributes unique info
âœ” VIF should ideally be < 5 (or < 10 max)
```

````python
import numpy as np
import pandas as pd
from statsmodels.formula.api import ols
import statsmodels.api as sm

# ============================
# Custom Boxâ€“Tidwell Function
# ============================

def box_tidwell_custom(y, X):
    """
    Runs Boxâ€“Tidwell transformation test for non-linearity.
    Returns lambda estimates and p-values.
    """

    lambdas = []
    pvalues = []
    cols = X.columns

    for col in cols:
        x = X[col]

        # Avoid log errors by replacing 0 or negative values
        x = np.where(x <= 0, 0.01, x)

        df_bt = pd.DataFrame({
            "y": y,
            col: x,
            col+"_logx": x * np.log(x)
        })

        # Fit model: y ~ x + x*log(x)
        formula = f"y ~ {col} + {col}_logx"
        model = ols(formula, data=df_bt).fit()

        # lambda = 1 + coefficient of x*log(x)
        lam = 1 + model.params[col+"_logx"]

        # p-value for non-linearity test
        p = model.pvalues[col+"_logx"]

        lambdas.append(lam)
        pvalues.append(p)

    return cols, np.array(lambdas), np.array(pvalues)


# ============================
# RUN BOXâ€“TIDWELL PIPELINE
# ============================

y = df["BodyFat"].astype(float)
X = df.drop(columns=["BodyFat"])

print("\nRunning Boxâ€“Tidwell Test...\n")

predictors, lambdas, pvalues = box_tidwell_custom(y, X)

print("========== BOXâ€“TIDWELL RESULTS ==========\n")

for col, lam, p in zip(predictors, lambdas, pvalues):
    print(f"{col:15} â†’ Î» = {lam:.4f}, p = {p:.4f}")


# ============================
# Apply transformations
# ============================

print("\n========== APPLYING TRANSFORMATIONS ==========\n")

for col, lam, p in zip(predictors, lambdas, pvalues):
    if p < 0.05:
        df[col + "_BT"] = (df[col].replace(0, 0.01)) ** lam
        print(f"Transformed: {col} â†’ {col}_BT (Î» = {lam:.4f})")
    else:
        print(f"No transform needed: {col}")

print("\nAll transformations applied successfully!")
df.tail()
````

````python
def fit_ols_and_metrics(X_subset, y, sigma2_full=None):
    X_const = sm.add_constant(X_subset)
    model = sm.OLS(y, X_const).fit()

    sse = np.sum(model.resid**2)
    r2 = model.rsquared
    adj_r2 = model.rsquared_adj
    aic = model.aic
    bic = model.bic

    p = int(model.df_model) + 1   # includes intercept

    if sigma2_full is not None:
        cp = sse / sigma2_full - (n - 2 * p)
    else:
        cp = np.nan

    return {
        "p": p - 1,
        "SSE": sse,
        "R_Squared": r2,
        "Adjusted_R_Squared": adj_r2,
        "Cp": cp,
        "AIC": aic,
        "BIC": bic
    }
````

````python
X_full_const = sm.add_constant(X)
full_model = sm.OLS(y, X_full_const).fit()

sse_full = np.sum(full_model.resid**2)
p_full = int(full_model.df_model) + 1
sigma2_full = sse_full / (n - p_full)

print("Full model ÏƒÂ²:", sigma2_full)
````

````python
results = []
max_features = len(feature_names)  # change if too large

model_id = 0

for k in range(1, max_features + 1):
    for combo in itertools.combinations(feature_names, k):
        X_subset = X[list(combo)]
        metrics = fit_ols_and_metrics(X_subset, y, sigma2_full=sigma2_full)
        metrics["Model_ID"] = model_id
        metrics["Variables_in_Model"] = list(combo)
        results.append(metrics)
        model_id += 1

results_df = pd.DataFrame(results)

results_df = results_df[
    ["Model_ID", "Variables_in_Model", "p", "SSE",
     "R_Squared", "Adjusted_R_Squared", "Cp", "AIC", "BIC"]
]

results_df.head()
````

````python
def forward_selection(X, y, criterion="aic"):
    selected = []
    remaining = list(X.columns)
    current_score, best_score = float("inf"), float("inf")

    while remaining:
        scores_with_candidates = []
        for candidate in remaining:
            model = sm.OLS(y, sm.add_constant(X[selected + [candidate]])).fit()
            if criterion == "aic":
                score = model.aic
            elif criterion == "adjr2":
                score = -model.rsquared_adj

            scores_with_candidates.append((score, candidate))

        scores_with_candidates.sort()
        best_new_score, best_candidate = scores_with_candidates[0]

        if best_new_score < best_score:
            remaining.remove(best_candidate)
            selected.append(best_candidate)
            best_score = best_new_score
        else:
            break

    final_model = sm.OLS(y, sm.add_constant(X[selected])).fit()
    return selected, final_model
````

````python
forward_vars, forward_model = forward_selection(X, y, criterion="aic")
print("Selected Variables (Forward):", forward_vars)
print(forward_model.summary())
````

````python
def backward_elimination(X, y, threshold=0.05):
    features = list(X.columns)
    
    while True:
        X_const = sm.add_constant(X[features])
        model = sm.OLS(y, X_const).fit()
        pvalues = model.pvalues.drop("const")

        worst_feature = pvalues.idxmax()
        worst_p = pvalues.max()

        if worst_p > threshold:
            features.remove(worst_feature)
        else:
            break
    
    final_model = sm.OLS(y, sm.add_constant(X[features])).fit()
    return features, final_model
````

````python

# ================================
# STEP 2 â€” Boxâ€“Tidwell Î» Values for X
# ================================
lambda_dict = {
    "Abdomen": -0.0601,
    "Hip": -1.2909,
    "Thigh": -1.5335,
    "Biceps": -3.5362,
    "Bmi": -0.4095   # we will calculate BMI first
}


# ================================
# STEP 3 â€” BMI Calculation
# ================================
def calculate_bmi(weight, height_cm):
    height_m = height_cm / 100
    return weight / (height_m ** 2)


# ================================
# STEP 4 â€” Apply Boxâ€“Tidwell Transformations
# ================================
def transform_features(user_input):
    transformed = {}

    for col, val in user_input.items():

        # Apply Boxâ€“Tidwell where needed
        if col in lambda_dict:
            lam = lambda_dict[col]
            if val <= 0:
                val = 0.01
            transformed[col + "_BT"] = val ** lam
        else:
            transformed[col] = val

    return transformed


# ================================
# STEP 5 â€” Prediction Function
# ================================
def predict_bodyfat(user_input):

    # ---- Auto compute BMI from weight + height ----
    if "Weight" in user_input and "Height" in user_input:
        user_input["Bmi"] = calculate_bmi(user_input["Weight"], user_input["Height"])

    # ---- Apply Boxâ€“Tidwell transformations ----
    transformed = transform_features(user_input)

    # ---- Convert to DataFrame ----
    X_input = pd.DataFrame([transformed])

    # ---- Align feature order with model ----
    X_input = X_input[model.params.index.drop("const")]

    # ---- Add constant ----
    X_const = sm.add_constant(X_input, has_constant='add')

    # ---- Predict ----
    pred = model.predict(X_const)
    return pred.values[0]


# ================================
# STEP 6 â€” SAMPLE USER INPUT
# ================================
user_input = {
    "Age": 28,
    "Neck": 37,
    "Knee": 40,
    "Forearm": 29,
    "Wrist": 17,
    "Abdomen": 91,
    "Hip": 98,
    "Thigh": 57,
    "Biceps": 32,
    "Weight": 72,   # kg
    "Height": 175   # cm
}

prediction = predict_bodyfat(user_input)
print("\nğŸ”¥ Predicted Body Fat % =", prediction)
````

````python
import numpy as np

def interpret_fitness_results(user_input, predicted_bf):

    weight = user_input["Weight"]
    height = user_input["Height"]
    height_m = height / 100

    # ===================== CORE METRICS ======================
    bmi = weight / (height_m ** 2)
    whr = user_input["Abdomen"] / user_input["Hip"]
    whtr = user_input["Abdomen"] / height
    fat_mass = (predicted_bf / 100) * weight
    lean_mass = weight - fat_mass
    mfr = lean_mass / fat_mass if fat_mass != 0 else np.inf  # Muscle-to-fat Ratio

    # ===================== ADVANCED METRICS ======================
    # Body Composition Indexes
    ffmi = lean_mass / (height_m ** 2)  # Fat Free Mass Index
    fmi = fat_mass / (height_m ** 2)    # Fat Mass Index
    mqi = lean_mass / weight            # Muscle Quality Index

    # Visceral fat proxy (scaled index)
    visceral_fat_score = (0.4 * whtr) + (0.4 * whr) + (0.2 * (predicted_bf / 30))

    # Strength/Muscle Structure Ratios
    strength_potential = user_input["Biceps"] / user_input["Forearm"]
    central_obesity_index = user_input["Abdomen"] / height
    android_gynoid_ratio = user_input["Abdomen"] / (user_input["Hip"] + user_input["Thigh"])

    # ===================== ENERGY METRICS ======================
    bmr = (10 * weight) + (6.25 * height) - (5 * user_input["Age"]) + 5
    tdee = bmr * 1.55  # default moderate activity

    protein_range = (1.6 * weight, 2.2 * weight)  # fat loss optimal range

    cal_goals = {
        "maintain": tdee,
        "slow_loss": tdee - 300,
        "fast_loss": tdee - 600,
        "lean_bulk": tdee + 300
    }

    # ===================== TARGET BODYFAT WEIGHTS ======================
    target_12 = lean_mass / 0.88
    target_15 = lean_mass / 0.85
    target_18 = lean_mass / 0.82

    # ===================== RISK SCORES ======================
    metabolic_risk = sum([
        bmi > 28,
        predicted_bf > 25,
        whr > 1.0
    ])

    # ===================== PRINT ALL RESULTS ======================
    print("\n================ FULL BODY ANALYSIS =================\n")

    print(f"Predicted Body Fat %: {predicted_bf:.2f}% ({classify_bodyfat(predicted_bf)})")
    print(f"BMI: {bmi:.2f} ({classify_bmi(bmi)})")
    print(f"WHR: {whr:.2f} ({classify_whr(whr)})")
    print(f"WHtR: {whtr:.2f} ({classify_wht(whtr)})")
    
    print("\n---- Body Composition ----")
    print(f"Lean Mass: {lean_mass:.2f} kg")
    print(f"Fat Mass: {fat_mass:.2f} kg")
    print(f"Muscle-to-Fat Ratio (MFR): {mfr:.2f}")

    print("\n---- Indexes (Fat & Muscle Quantification) ----")
    print(f"FFMI (Muscle Mass): {ffmi:.2f}")
    print(f"FMI (Fat Mass Relative to Height): {fmi:.2f}")
    print(f"Muscle Quality Index (MQI): {mqi:.2f}")

    print("\n---- Body Geometry ----")
    print(f"Central Obesity Index: {central_obesity_index:.2f}")
    print(f"Android/Gynoid Ratio: {android_gynoid_ratio:.3f}")
    print(f"Strength Potential Index (Biceps/Forearm): {strength_potential:.2f}")

    print("\n---- Metabolic & Energy Metrics ----")
    print(f"BMR: {bmr:.0f} kcal/day")
    print(f"TDEE (Estimated): {tdee:.0f} kcal/day")
    print(f"Protein Target Range: {protein_range[0]:.0f}â€“{protein_range[1]:.0f} g/day")

    print("\n---- Calorie Goals ----")
    for k, v in cal_goals.items():
        print(f"{k}: {v:.0f} kcal")

    print("\n---- Target Bodyweights ----")
    print(f"Weight at 12% BF: {target_12:.2f} kg")
    print(f"Weight at 15% BF: {target_15:.2f} kg")
    print(f"Weight at 18% BF: {target_18:.2f} kg")

    print("\n---- Risk Scores ----")
    print(f"Metabolic Syndrome Risk (0â€“3): {metabolic_risk}")
    print(f"Visceral Fat Score (0â€“10 est.): {visceral_fat_score:.2f}")

    print("\n==============================================================\n")
````
